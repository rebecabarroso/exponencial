# Distribuições contínuas: Distribuição Exponencial

**Definição:** Se caracteriza por ter uma função de taxa
de falha constante. A distribuição exponencial é a única com esta
propriedade.

image.png

---

**Exemplo**: tempo de vida de certos produtos.

β: 0.06

```python
import pandas as pd
import math
from scipy import stats
import numpy as np
from scipy.stats import expon
import matplotlib.pyplot as plt
import sympy as sp
from scipy.optimize import curve_fit
```

```python
# criando varáveis aleatórias
beta = 0.06
valor_n = expon.rvs(scale=1/beta, size=80)
print('variáveis aleatórias:', valor_n)
```

```python
# gráfico
plt.hist(valor_n, bins=15, density=True,alpha = 0.6, color='b', label='Histograma dos Dados')
```

## Função Geradora de Momentos
image.png

```python
# calculando fgm
valor_t=0.02
fgm = beta/(beta-valor_t)
print('As funções geradoras de momentos, para todos t valores são:', fgm)
```

#Esperança
image.png

```python
# calculando a esperança
# primeira derivada da fgm
esperanca=1/beta
print('A esperança é:', esperanca)
```

#Esperança ao quadrado
image.pngimage.png

```python
# calculando a esperança ao quadrado
# segunda derivada da fgm
esp_quad = 2/(beta**2)
print('A esperança ao quadrado é:', esp_quad)
```

#Variância
image.pngimage.png

```python
var = 1/(beta**2)
print('A variância da distribuição é:', var)
```
